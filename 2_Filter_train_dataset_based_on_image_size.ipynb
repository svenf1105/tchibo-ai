## Notebook overview:
- With this notebook we can modify our train/test data by fitering out pictures with specific dimensions
- The notebook will move the picture and the belonging annotation into a new folder

```python
import os
import warnings
import shutil
import cv2

#create a list of pcitures filepaths
image_dir= '/content/drive/MyDrive/training_demo/images/drive-download-20220131T190918Z-001/test_bohne' 
files = os.listdir(image_dir)
#print(files)
image_list = []
for file in files:
  if file.endswith("jpg"):
    image_list.append(image_dir +"/" + file)
print(image_list)

for image_path in image_list:
  img = cv2.imread(image_path)
  img.shape
  #move train data + annotations to other directory based on the conditions
  if int(img.shape[0]) >= 512 and int(img.shape[1]) >= 512:
    try:
      base=os.path.basename(image_path)
      annotation_path = image_dir +'/'+os.path.splitext(base)[0]+".xml"
      #move image to folder
      source_img = image_path
      destination = "/content/drive/MyDrive/training_demo/images/test_bohne_512"
      shutil.move(source_img,destination)
      #move the annotation to folder 
      source_ann = annotation_path
      shutil.move(source_ann,destination)
      print(image_path +" was moved to: "+ destination)
    except:
      pass
```

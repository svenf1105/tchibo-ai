```python
import pandas as pd
import requests
import random
```

```python
#Liste mit URLs einspielen
df = pd.read_csv('tchibo_community_images.csv')
df.head()
```

```python
#Liste mit gewnschten Bildern festlegen, in diesem Beispiel 3000 Random-Bilder.
df_list = df.image.tolist()
df_random3000 = random.sample(df_list, 3000)
#Check, ob die gewnschte Anzahl an Bildern in der Liste ist
len(df_random3000)
```

```python
# Funktionierende Bilder werden heruntergeladen in den aktuellen Ordner, nicht funktionierende (404 Error) kommen in eine Liste

broken_images = []

for img in df_random3000:
    # split the file based upon / and extract the last split within the python list below:
    file_name = img.split('/')[-1]
    print(f"This is the file name: {file_name}")
    # request to the image URL:
    r = requests.get(img, stream=True)
    # status code 200 = ok. 404 = nicht ok:
    if r.status_code == 200:
        # write the data to a file as binary:
        with open(file_name, 'wb') as f:
            for chunk in r:
                f.write(chunk)
    else:
        # We will write all of the images back to the broken_images list:
        broken_images.append(img)
```

```python
#Check, ob welche der Bilder nicht funktionieren
broken_images
```

```python
# Links der Bilder werden in eine CSV gespeichert, um spter einen Abgleich auf Duplikate mit anderen Listen machen zu knnen.
dict = {'url': df_random3000}  
       
df2 = pd.DataFrame(dict) 
    
# saving the dataframe 
df2.to_csv('Random300_fr_Testset.csv') 
```
